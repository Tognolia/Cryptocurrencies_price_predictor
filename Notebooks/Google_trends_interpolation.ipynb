## Google trends interpolation
- In this passage i am going to get the data in csv from google trends page in weekly format
- Create a new list of daily trends through interpolation method
- export the file in order to later use it in the model


```python
# Import libraries
import pandas as pd
import datetime
```

```python
# read the csv
gtrends = pd.read_csv('googletrends_final.csv', sep=';')
gtrends.head()
```

```python
len(gtrends)
```

```python
# creating a new df where will append the gtrends
new_df = pd.DataFrame()
new_df["Date"] = pd.date_range(start='2017-08-13', end='2021-05-16')
new_df
```

```python
list_trends = gtrends['Score'].to_list()
```

```python
# creating the list to interpolate
list = []
for i in list_trends:
    list.append(i)
    list.append(None)
    list.append(None)
    list.append(None)
    list.append(None)
    list.append(None)
    list.append(None)
list   
```

```python
#create a list witht the same length
list = list[:1373]
len(list)
```

```python
# new list obtained
new_df['trends'] = list
new_df
```

```python
# linearinterpolation method to infer the missing trends values
new_df["trends"] = new_df["trends"].interpolate(method = "linear")
new_df
```

```python
# adapt to the lenght of the thereum price hist.
new_df = new_df.iloc[4:1368]
```

```python
new_df.to_csv('gtrends_interpolated.csv')
```
